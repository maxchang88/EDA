```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/EDA/assignment/data/ForeignGifts_edu.csv')

# 2
plt.hist(df['Foreign Gift Amount'], bins=50, edgecolor='black')
plt.xlabel('Foreign Gift Amount')
plt.ylabel('Frequency')
plt.title('Histogram of Foreign Gift Amount')
plt.show()

df['Foreign Gift Amount'].describe()

# The distribution is very right skewed containing very few large donations and
# many smaller donations. Something odd about this variable is that there are a
# decent handful of observations with negative donation amounts. Despite the
# abberations of a few outliars this histogram is expected.
```

```python
# 3

# Value counts table
gift_type_counts = df['Gift Type'].value_counts()
# Getting the proportions of each gift type
gift_type_proportions = df['Gift Type'].value_counts(normalize=True)

print("Count of each gift type:")
print(gift_type_counts)
print("\nProportion of each gift type:")
print(gift_type_proportions)
```

```python
# 4

# Taking log
df['log_foreign_gift'] = np.log(df['Foreign Gift Amount'] + 1)

# Plot the KDE
sns.kdeplot(data=df, x='log_foreign_gift', fill=True)
plt.title('KDE Plot of Log(Foreign Gift Amount)')
plt.xlabel('log(Foreign Gift Amount + 1)')
plt.ylabel('Density')
plt.show()

# Plot conditional KDE
sns.kdeplot(data=df, x='log_foreign_gift', hue='Gift Type', fill=True)
plt.title('KDE Plot of Log(Foreign Gift Amount) by Gift Type')
plt.xlabel('log(Foreign Gift Amount + 1)')
plt.ylabel('Density')
plt.show()

# The regular kernal density plot closely mirrors the density plot of the contract
# gift type
```

```python
# 5

# Group by country and count the number of gifts
country_gift_counts = df.groupby('Country of Giftor')['Foreign Gift Amount'].count()
# Sort descending and get top 15
top15_by_count = country_gift_counts.sort_values(ascending=False).head(15)

print("Top 15 countries by number of gifts:")
print(top15_by_count)

# Group by country and sum the gift amounts
country_gift_sums = df.groupby('Country of Giftor')['Foreign Gift Amount'].sum()
# Sort descending and get top 15
top15_by_sum = country_gift_sums.sort_values(ascending=False).head(15)

print("\nTop 15 countries by total gift amount:")
print(top15_by_sum)
```

```python
# 6

# Group by institution and sum the gift amounts
institutions_sum = df.groupby('Institution Name')['Foreign Gift Amount'].sum()
# Sort in descending order and pick the top 15
top_15_institutions = institutions_sum.sort_values(ascending=False).head(15)

print("Top 15 Institutions by Total Amount Received:")
print(top_15_institutions)

plt.hist(institutions_sum, bins=30, edgecolor='black', color='skyblue')
plt.xlabel('Total Gift Amount Received')
plt.ylabel('Number of Institutions')
plt.title('Histogram of Total Amount Received by Institutions')
plt.show()
```

```python
# 7

# Group the data by giftor and sum the total amounts
giftor_sums = df.groupby('Giftor Name')['Foreign Gift Amount'].sum()
# Sort in descending order to see who gives the most overall
top_giftors = giftor_sums.sort_values(ascending=False)

print("Giftors providing the most money (in total):")
print(top_giftors.head(15))
```
